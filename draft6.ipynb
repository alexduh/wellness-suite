```python
from typing import Any, Dict

from google.adk.agents import Agent, LlmAgent, SequentialAgent, ParallelAgent
from google.adk.apps.app import App, EventsCompactionConfig
from google.adk.models.google_llm import Gemini
from google.adk.sessions import DatabaseSessionService
from google.adk.sessions import InMemorySessionService
from google.adk.runners import Runner, InMemoryRunner
from google.adk.plugins.logging_plugin import (LoggingPlugin)
from google.adk.tools import AgentTool, FunctionTool, google_search
from google.adk.tools.tool_context import ToolContext
from google.adk.code_executors import BuiltInCodeExecutor
from google.genai import types

print("Sessions and context engineering (3a) and MAS (1b) and Tools (2a) ADK imports")
```

```python
APP_NAME = "Wellness Suite"
USER_ID = "default"
SESSION = "default"

MODEL_NAME = "gemini-2.5-flash-lite"
```

```python
retry_config = types.HttpRetryOptions(
    attempts=3,
    exp_base=7, # Delay multiplier
    initial_delay=1, # initial delay before first retry (seconds)
    http_status_codes=[429, 500, 503, 504] # Retry on these HTTP errors
)
```

```python
# Create a SQLite database with DatabaseSessionService
db_url = "sqlite:///agentic_wellness_data.db"

# Set up Session Management
session_service = DatabaseSessionService(db_url=db_url)
```

```python
# Helper functions for running a session
async def run_session(
    runner_instance: Runner,
    user_queries: list[str] | str = None,
    session_name: str = "default",
):
    print(f"\n ### Session: {session_name}")

    # Get app name from the Runner
    app_name = runner_instance.app_name

    # Attempt to create a new session or retrieve an existing one
    try:
        session = await session_service.create_session(
            app_name=app_name, user_id=USER_ID, session_id=session_name
        )
    except:
        session = await session_service.get_session(
            app_name=app_name, user_id=USER_ID, session_id=session_name
        )

    # Process queries if provided
    if user_queries:
        # Convert single query to list for uniform processing
        if type(user_queries) == str:
            user_queries = [user_queries]

        # Process each query in the list sequentially
        for query in user_queries:
            print(f"\nUser > {query}")

            # Convert the query string to the ADK content format
            query = types.Content(role="user", parts=[types.Part(text=query)])

            # Stream the agent's response asynchronously
            async for event in runner_instance.run_async(
                user_id=USER_ID, session_id=session.id, new_message=query
            ):
                # Check if the event contains valid content
                if event.content and event.content.parts:
                    # Filter out empty or "None" responses before printing
                    if (
                        event.content.parts[0].text != "None"
                        and event.content.parts[0].text
                    ):
                        print(f"{MODEL_NAME} > ", event.content.parts[0].text)
    else:
        print("No queries!")
```

```python
# Session state management tools
USER_NAME_SCOPE_LEVELS = ("temp", "user", "app")

# 'user:' prefix indicates the following is user-specific data
def save_userinfo(
    tool_context: ToolContext, user_name: str, user_age: str, user_height: str, user_weight: str, user_gender: str,
    user_dietary_preferences: str, user_current_fitness_level: str, user_fitness_goals: str, user_injury_history: str,
) -> Dict[str, Any]:
    """
    Tool to record and save user personal wellness information in session state.

    Args:
        user_name: The username to store in session state
        user_age: The user's age to store in session state
        user_height: The user's height to store in session state
        user_weight: The user's weight to store in session state
        user_gender: The user's gender to store in session state
        user_dietary_preferences: The user's dietary preferences to store in session state
        user_current_fitness_level: The user's current fitness level to store in session state
        user_fitness_goals: The user's fitness goals to store in session state
        user_injury_history: The user's injury history to store in session state
    """
    # Write to session state using the 'user:' prefix for user data
    tool_context.state["user:name"] = user_name
    tool_context.state["user:age"] = user_age
    tool_context.state["user:height"] = user_height
    tool_context.state["user:weight"] = user_weight
    tool_context.state["user:gender"] = user_gender
    tool_context.state["user:dietary_preferences"] = user_dietary_preferences
    tool_context.state["user:current_fitness_level"] = user_current_fitness_level
    tool_context.state["user:fitness_goals"] = user_fitness_goals
    tool_context.state["user:injury_history"] = user_injury_history

    return {"status": "success"}

# tools can read from session state. example:
def retrieve_userinfo(tool_context: ToolContext) -> Dict[str, Any]:
    """
    Tool to retrieve user name from session state.
    """

    user_name = tool_context.state.get("user:name", "User name not found")
    user_age = tool_context.state.get("user:age", "User age not found")
    user_height = tool_context.state.get("user:height", "User height not found")
    user_weight = tool_context.state.get("user:weight", "User weight not found")
    user_gender = tool_context.state.get("user:gender", "User gender not found")
    user_dietary_preferences = tool_context.state.get("user:dietary_preferences", "User dietary preferences not found")
    user_current_fitness_level = tool_context.state.get("user:current_fitness_level", "User current fitness level not found")
    user_fitness_goals = tool_context.state.get("user:fitness_goals", "User fitness goals not found")
    user_injury_history = tool_context.state.get("user:injury_history", "User injury history not found")

    return {"status": "success", "user_name": user_name, "age": user_age, "height": user_height, "weight": user_weight,
            "gender": user_gender, "dietary_preferences": user_dietary_preferences, 
            "current_fitness_level": user_current_fitness_level,
            "fitness_goals": user_fitness_goals, "injury_history": user_injury_history}
```

```python
# Specialist Agents

# Fitness Trainer Agent: focuses on fitness
fitness_trainer_agent = LlmAgent(
    name="FitnessTrainer",
    model=Gemini(
        model=MODEL_NAME,
        retry_options=retry_config
    ),
    instruction="""You are an expert fitness trainer with deep knowledge of human physiology.
    Please provide a detailed fitness training plan for the user
    that is customized for the user's personal wellness information (use the `retrieve_userinfo` tool to obtain this)
    """,
    tools=[retrieve_userinfo],
    output_key="fitness_plan"
)

# Nutritionist Agent: focuses on nutrition
nutritionist_agent = LlmAgent(
    name="Nutritionist",
    model=Gemini(
        model=MODEL_NAME,
        retry_options=retry_config
    ),
    instruction="""You are an expert nutritionist with deep knowledge of diet and human metabolism.
    Please provide a detailed nutrition plan for the user that is customized
    for the user's personal wellness information (use the `retrieve_userinfo` tool to obtain this)
    Ensure that the nutrition plan meets the macronutrient and micronutrient needs to achieve their goals
    """,
    tools=[retrieve_userinfo],
    output_key="nutrition_plan"
)
```

```python
# Aggregator Agent

aggregator_agent = LlmAgent(
    name="WellnessAggregator",
    model=Gemini(
        model=MODEL_NAME,
        retry_options=retry_config
    ),
    instruction="""Combine the fitness plan and nutrition plan into a single wellness plan tailored for the user.
    Focus on the synergy between the fitness and nutrition aspects to maximize the user's health and wellness outcomes.

    **Fitness Plan:**
    {fitness_plan}

    **Nutrition Plan:**
    {nutrition_plan}

    Your final wellness plan should seamlessly integrate both components and provide clear, actionable recommendations.
    """,
    tools=[],
    output_key="wellness_plan"
)
```

```python
# Parallel Agent to run both specialists simultaneously
parallel_wellness_team = ParallelAgent(
    name="ParallelWellnessTeam",
    sub_agents=[fitness_trainer_agent, nutritionist_agent],
)
```

```python
# Sequential Agent to run the parallel team first, then run the aggregator
sequential_wellness = SequentialAgent(
    name="WellnessSystem",
    sub_agents=[parallel_wellness_team, aggregator_agent],
)
```

```python
# Intake Agent

intake_agent = LlmAgent(
    name="WellnessIntakeAgent",
    model=Gemini(
        model_name=MODEL_NAME,
        retry_options=retry_config
    ),
    instruction="""You are an expert wellness assistant agent with concise yet kind verbiage.
    Ensure that the user provides necessary personal information for fitness and nutrition advice by asking questions about their health.
    Use the `retrieve_userinfo` tool to check what information you currently have and lack to determine which questions to ask.
    Craft your questions to collect the most essential information.
    """,
    tools=[retrieve_userinfo],
    output_key="intake_questions",
)
```

```python
# Centralized root Wellness Coordinator Agent to provide holistic wellness plan

root_agent = LlmAgent(
    name="WellnessCoordinatorAgent",
    model=Gemini(
        model_name=MODEL_NAME,
        retry_options=retry_config
    ),
    instruction="""You are the Wellness Coordinator Agent.
    Your goal is to assist users in improving their overall wellness by orchestrating the following workflow.
    1. First, ascertain whether or not you have the user's necessary personal information to provide customized fitness and nutrition advice. Use the `retrieve_userinfo` tool to do so.
    2. If you lack any information, call the `WellnessIntakeAgent` tool to craft questions to obtain the necessary personal information.
    3. Use the `save_userinfo` tool to save personal information about the user over the course of your conversation.
    4. Once you have sufficient information, call the `WellnessSystem` tool to create a wellness plan for the user.
    5. Present the results of the wellness plan to the user.
    """,
    tools=[
        save_userinfo,
        retrieve_userinfo,
        AgentTool(intake_agent),
        AgentTool(sequential_wellness)
    ],
)
```

```python
# App with Events Compaction enabled
wellness_app_compacting = App(
    name="wellness_app_compacting",
    root_agent=root_agent,
    plugins=[LoggingPlugin()], 
    events_compaction_config=EventsCompactionConfig(
        compaction_interval=2, # Trigger compaction every 2 invocations
        overlap_size=1, # Keep 1 previous turn for context
    ),
)
```

```python
wellness_runner_compacting = Runner(app=wellness_app_compacting, session_service=session_service)
```
